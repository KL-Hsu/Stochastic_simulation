# I. Import requirements

```python
import numpy as np 
import random
from matplotlib import pyplot as plt
import time
```

```python
def timer(func):
    def wrapper_timer(*args, **kwarg):
        start = time.time()
        
        values = func(*args, **kwarg)
        
        end = time.time()
        run_time = end - start
        print('Runtime: ', run_time, 's')
        
        return values
    return wrapper_timer
```

# II. Monte Carlo Simulation

```python
inside = 0
n = int(1e5)

inside_x = []
inside_y = []
outside_x = []
outside_y = []

for i in range(n):
    x=random.random()
    y=random.random()
    if ((x-0.5)**2+(y-0.5)**2)<=0.25:
        inside+=1
        inside_x.append(x)
        inside_y.append(y)
    else:
        outside_x.append(x)
        outside_y.append(y)       

pi=4*inside/n
print(pi)

figure, ax = plt.subplots(figsize=(6, 6))
plt.plot(inside_x, inside_y, '.', markersize=1)
plt.plot(outside_x, outside_y, '.', markersize=1)
plt.show()
```

```python
inside = 0
n = int(1e5)

inside_x = []
inside_y = []
outside_x = []
outside_y = []

for i in range(n):
    x=random.random()
    y=random.random()
    if (x*x + y*y) <= 1:
        inside+=1
        inside_x.append(x)
        inside_y.append(y)
    else:
        outside_x.append(x)
        outside_y.append(y)       

pi=4*inside/n
print(pi)

figure, ax = plt.subplots(figsize=(6, 6))
plt.plot(inside_x, inside_y, '.', markersize=1)
plt.plot(outside_x, outside_y, '.', markersize=1)
plt.show()
```

```python
@timer 
def Estimate_Pi_by_MonteCarlo(number_per_each=int(1e5), times=1000): 
    
    Pi_ls = []
    for i in range(times):
        inside = 0
        for j in range(number_per_each):
            x = random.random()
            y = random.random()
            if (x*x + y*y) <= 1:
                inside += 1 
        Pi = 4*inside/n
        Pi_ls.append(Pi)
    
    return Pi_ls 
```

```python
Estimated_Pi_ls = Estimate_Pi_by_MonteCarlo(times=5000)
```

```python
plt.hist(Estimated_Pi_ls, bins=100, density=True)
plt.title('Estimate Pi by Monte Carlo method')
plt.ylabel('Counts')
plt.xlabel('Estimated Pi')
plt.show()

mean = np.array(Estimated_Pi_ls).mean()
std = np.array(Estimated_Pi_ls).std()

print('mean: ', mean)
print('standard deviation: ', std)
```

# III. The Gillespie's direct method

   1. Generate a random value $\tau$ according to $P(\tau)$ 
   2. Generate a random integer $\mu$ according to $P(\tau, \mu)$ 

## 1. Generate a random value by a distribution

```python
#Time interval tau
tau = np.linspace(1, 100, 1000)
distribution = np.e**(-0.2*tau)

#Validate if the random.choices based on the distribution
result = []
for i in range(10000):
    s = random.choices(tau, distribution)
    result.append(s[0])  
    
plt.hist(result, bins=100)
plt.show()
```

```python
def generate_distribution(c_ls, tau):
    a = 0
    for i in c_ls:
        a += i
    return a*np.e**(-a*tau)
```

```python
tau = np.linspace(0.01, 1, 1000)

for i in range(5):
    P_ls = []
    P = 0
    t = 0
    k = 0.5

    c1 = 100
    c2 = P*k
    t_end = 100
    P_ls.append([t, P])

    while t < t_end:
        c1 = c1
        c2 = P*k
        distribution = gen_distribution([c1, c2], tau)
        t_interval = random.choices(tau, distribution)
        delta = random.choices([1, -1], [c1, c2]) 
        t += t_interval[0]
        P += delta[0]

        P_ls.append([t, P])

    time_point = [x[0] for x in P_ls] 
    number = [x[1] for x in P_ls]

    plt.plot(time_point, number)
    
plt.show()
```
